
NUMPY




```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)
```

```python
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)
zeros_arr = np.zeros(5)
ones_arr = np.ones((2, 3))
range_arr = np.arange(1, 10, 2)
transpose_arr = arr.transpose()
print("Zeros Array:", zeros_arr)
print("Ones Array:", ones_arr)
print("Range Array:", range_arr)
print("Transpose Array: ", transpose_arr)
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
addition = arr1 + arr2
subtraction = arr1 - arr2
multiplication = arr1 * arr2
division = arr1 / arr2
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
```

```python
arr_int = np.array([1, 2, 3], dtype=int)
arr_float = np.array([1.0, 2.5, 3.2], dtype=float)
print("Integer Array:", arr_int)
print("Float Array:", arr_float)
print("Data type of arr_int:", arr_int.dtype)
print("Data type of arr_float:", arr_float.dtype)
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:", arr)
print("Dimensions:", arr.ndim)
print("Shape:", arr.shape)
```
